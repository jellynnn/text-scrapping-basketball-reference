```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.basketball-reference.com/'

response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    headlines = soup.find_all('h2')

    for i, headline in enumerate(headlines):
        print(f"{i + 1}. {headline.text.strip()}")
else:
    print(f"Gagal mengunduh halaman. Status code: {response.status_code}")
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2024_per_game.html'

response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find('table', {'id': 'per_game_stats'})

    headers = []
    for th in table.findAll('th'):
        if th.get('data-stat') != 'rank':
            headers.append(th.text)

    data = []
    for row in table.findAll('tr')[1:]:
        cols = row.findAll('td')
        if cols:
            data.append([col.text for col in cols])

    print(headers)
    for player in data:
        print(player)
else:
    print("Error fetching the page")

```

```python
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'id': 'per_game_stats'})

headers = [th.text for th in table.find('thead').find_all('th')][1:]

rows = table.find('tbody').find_all('tr')
player_data = []
for row in rows:
    cols = row.find_all('td')
    if cols:
        player_data.append([col.text for col in cols])

df = pd.DataFrame(player_data, columns=headers)

print(df.head())
```
